```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(X= df.drop('Private',axis=1))
```

```python
kmeans.cluster_centers_
```

```python
def converter(cols):
    if cols == 'Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
dummies = pd.get_dummies(data=df['Private'],drop_first=True)
```

```python
df_new = pd.concat([df,dummies],axis=1)
```

```python
df_new.head()
```

```python
df_new['Cluster'] = df_new['Yes']
```

```python
df_new.drop('Yes',axis=1,inplace=True)
```
